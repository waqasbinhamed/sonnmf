```python
import numpy as np
from sonnmf.expanded import non_neg

W = np.random.uniform(-1, 1, (5, 5))
nnW = non_neg(W)
```

```python
np.maximum(0, W) == nnW
```

```python
from sonnmf.expanded import calculate_fscore as cf1
from sonnmf.gpt import calculate_fscore as cf2

m, n, r = 50, 30, 8
M = np.random.random((m, n))
W = np.random.random((m, r))
H = np.random.random((r, n))

cf1(M, W, H) == cf2(M, W, H)
```

```python
from sonnmf.expanded import calculate_gscore as cg1
from sonnmf.gpt import calculate_gscore as cg2

# W = np.random.uniform(-1, 1, (m, r))
cg1(W) == cg2(W)
```

```python
from sonnmf.expanded import calculate_hscore as ch1
from sonnmf.gpt import calculate_hscore as ch2


ch1(W) == ch2(W)
```

```python
from sonnmf.expanded import prox as p1
from sonnmf.gpt import prox as p2


p1(0.3, W[:, 0:1], W[:, 1:2]) == p2(0.3, W[:, 0:1], W[:, 1:2])
```

```python
from sonnmf.expanded import proj_simplex as ps1
from sonnmf.gpt import proj_simplex as ps2

ps1(W[:, 0:1]) == ps2(W[:, 0:1])
```

```python
from sonnmf.expanded import sonnmf_with_penalty as nmf1
from sonnmf.gpt import nmf_with_penalty as nmf2

nmf1(M, W.copy(), H.copy(), lam=1, gamma=1, itermax=10, verbose=True)
```

```python
nmf2(M, W.copy(), H.copy(), lam=1, gamma=1, itermax=10, verbose=True)
```
